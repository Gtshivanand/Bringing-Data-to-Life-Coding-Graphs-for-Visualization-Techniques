```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Data preparation
data = {
    "Name": ["Aditya", "Bhavika", "Chirag", "Deepa", "Esha", "Farhan", "Gita", "Hitesh", "Ila", "Jai",
             "Kavya", "Lalit", "Mira", "Nikhil", "Om", "Priya", "Raj", "Sunita", "Tarun", "Usha"],
    "Age": [20, 19, 21, 22, 20, 19, 21, 20, 18, 22, 19, 20, 21, 22, 19, 18, 21, 20, 19, 21],
    "Course": ["BCA", "BBA", "Bcom", "BCA", "BBA", "BCA", "Bcom", "BCA", "BBA", "Bcom", "BCA", "BBA", "Bcom", "BCA", "BBA", "Bcom", "BCA", "BBA", "Bcom", "BCA"],
    "Semester": [4, 2, 5, 6, 3, 1, 6, 4, 1, 5, 2, 3, 6, 5, 2, 1, 5, 3, 4, 6],
    "Attendance_Percentage": [88, np.nan, 92, 94, np.nan, 85, 78, 101, 87, 85, np.nan, 95, 82, 93, 75, 95, 90, np.nan, 88, 92],
    "Height_cm": [172, 158, 180, 160, 70, 165, 170, 165, 190, 180, 159, 175, np.nan, 174, 168, 154, 182, 160, 80, 164],
    "Weight_kg": [70, 50, 80, np.nan, 54, 65, 75, 170, 48, 82, 110, 80, 70, 76, 60, 55, 83, 120, 40, 59]
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
import warnings 
warnings.filterwarnings( "ignore")
```

```python
# Filling missing values with the mean for simplicity
df['Attendance_Percentage'].fillna(df['Attendance_Percentage'].mean(), inplace=True)
df['Height_cm'].fillna(df['Height_cm'].mean(), inplace=True)
df['Weight_kg'].fillna(df['Weight_kg'].mean(), inplace=True)
```

```python
# 1. Histogram
plt.figure(figsize=(12, 6))
sns.histplot(df['Age'], kde=True, bins=10, color='blue')
plt.title('Age Distribution')
plt.show()
```

```python
# 2. Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Course', y='Weight_kg', data=df)
plt.title('Boxplot of Weight by Course')
plt.show()
```

```python
# 3. Time Series Plot (interpreting Semester as time)
plt.figure(figsize=(12, 6))
sns.lineplot(x='Semester', y='Attendance_Percentage', data=df, marker='o')
plt.title('Attendance Percentage over Semesters')
plt.show()
```

```python
# 4. Heatmap (correlation matrix)
corr_matrix = df[['Age', 'Attendance_Percentage', 'Height_cm', 'Weight_kg']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 5. Scatter Plot (Age vs Weight)
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Age', y='Weight_kg', data=df, hue='Course')
plt.title('Scatter Plot of Age vs Weight')
plt.show()
```

```python
# 6. Line Chart (Attendance Percentage by Age)
plt.figure(figsize=(12, 6))
sns.lineplot(x='Age', y='Attendance_Percentage', data=df, marker='o', color='green')
plt.title('Attendance Percentage by Age')
plt.show()
```
